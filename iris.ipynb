# Iris Dataset

This notebook explores the popular Iris dataset containing features distinguishing various varities of the flower Iris.

```python
from sklearn import datasets

iris = datasets.load_iris()

print iris.DESCR
```

## Visualization

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

X = iris.data
y = iris.target
```

### Sepal Length v.s Sepal Width

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length v.s Sepal Width')
```

### Petal Length v.s Petal Width

```python
plt.scatter(X[:, 2], X[:, 3], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Width v.s Petal Length')
```
